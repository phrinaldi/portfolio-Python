```python
# importar bibliotecas:

## pandas = para trabalhar com as tabelas
import pandas

## IPython = para "envio" de emails
from IPython.display import HTML

## pathlib = para manuseio de pastas no computador
import pathlib
```

```python
# importar bases de dados:

## obs: importante deixar, neste caso, o arquivo python e a pasta "Bases de Dados" dentro da mesma pasta
emails = pandas.read_excel(r'Bases de Dados\Emails.xlsx')
lojas = pandas.read_csv(r'Bases de Dados\Lojas.csv', encoding='latin-1', sep=';')
vendas = pandas.read_excel(r'Bases de Dados\Vendas.xlsx')
display(emails)
display(lojas)
display(vendas)
```

```python
# incluir nome da loja em vendas, ou seja, mesclar as tabelas vendas e lojas. Isso é feito para facilitar o processo de 
# criação da tabela para cada loja

vendas = vendas.merge(lojas, on='ID Loja')
display(vendas)
```

```python
# criar uma tabela para cada loja
## ideia: fazer um dicionario em que cada item é uma tabela de uma loja. Ou seja, para cada loja na tabela lojas, cria-se uma
## chave (para a loja) e o valor é a tabela filtrada da loja. Assim, conseguimos fazer um 'For' para percorrer a coluna Loja e
## armazenar cada tabela nesse dicionario.

dicionario_lojas = {}
for loja in lojas['Loja']: # para cada loja na coluna Loja da tabela lojas...
    dicionario_lojas[loja] = vendas.loc[vendas['Loja']==loja,:] # ...cria uma chave para a loja e o valor é a tabela de vendas filtrada na loja

## display de dois exemplos para ver se deu certo
display(dicionario_lojas['Iguatemi Esplanada'])
display(dicionario_lojas['Norte Shopping'])
```

```python
# definir o dia
## data é um numero. Logo, o maior valor é a data mais recente. Pegamos o maior valor para saber o ultimo dia.

dia_indicador = vendas['Data'].max()
print(dia_indicador.day)
```

```python
# identificar se a pasta ja existe

## o nome das pastas será o nome das lojas. POR ISSO, VERIFICAREI SE JÁ EXISTEM PASTAS COM OS NOMES DAS LOJAS

## colocar o caminho numa variavel
caminho_backup = pathlib.Path(r'Backup Arquivos Lojas')

1## fazer uma lista das pastas que já existem (importante para o passo do FOR)
arquivos_pasta_backup = caminho_backup.iterdir()

2## fazer uma lista vazia para receber os nomes das pastas (importante para o proximo passo FOR)
lista_nomes_backup = []

2## para cada pasta na pasta de backup, adicione o nome da pasta na lista de nomes
for arquivo in arquivos_pasta_backup:
    lista_nomes_backup.append(arquivo.name)

3## verificar se já tem o nome da loja na lista, pq se não tiver, é pq é preciso criar uma pasta com esse nome
for loja in dicionario_lojas: # para cada chave no diconario
    if loja not in lista_nomes_backup: # se nao tiver a chave na lista de nomes
        nova_pasta = caminho_backup / loja # cria-se o caminho da nova pasta com o nome da loja
        nova_pasta.mkdir() # cria-se a nova pasta
    
    ## salvar dentro da pasta
    ## já tendo a pasta, criamos aqui o caminho do arquivo com seu nome (cria nome e caminho)
    nome_arquivo = f'{dia_indicador.day}_{dia_indicador.month}_{loja}.xlsx'
    local_arquivo = caminho_backup / loja / nome_arquivo
    
    ## pegar a tabela da loja por meio do dicionario e transformar em um arquivo excel (leva o excel para o caminho)
    dicionario_lojas[loja].to_excel(local_arquivo)
```

```python
# definicao de metas

meta_faturamento_dia = 1000
meta_faturamento_ano = 1650000
meta_qteprodutos_dia = 4
meta_qteprodutos_ano = 120
meta_ticketmedio_dia = 500
meta_ticketmedio_ano = 500
```

```python
for loja in dicionario_lojas:

    # Criar 2 tabelas para fazer os indicadores (acumulado do ano, filtrando só a loja; e do dia, filtrando loja e dia)
    vendas_loja = dicionario_lojas[loja]
    vendas_loja_dia = vendas_loja.loc[vendas_loja['Data']==dia_indicador, :]


    # faturamento do ano e do dia (somar a coluna valor final)
    faturamento_ano = vendas_loja['Valor Final'].sum()
    #print(faturamento_ano)
    faturamento_dia = vendas_loja_dia['Valor Final'].sum()
    #print(faturamento_dia)

    # diversidade de produtos (metodo unique para dar os valores unicos e calcular a len da lista gerada pelo unique)
    qte_produtos_ano = len(vendas_loja['Produto'].unique())
    #print(qte_produtos_ano)
    qte_produtos_dia = len(vendas_loja_dia['Produto'].unique())
    #print(qte_produtos_dia)

    # ticket medio
    valor_por_venda = vendas_loja.groupby('Código Venda').sum()
    
    ticket_medio_ano = valor_por_venda['Valor Final'].mean()
    #print(ticket_medio_ano)
    valor_por_venda_dia = vendas_loja_dia.groupby('Código Venda').sum()
    
    ticket_medio_dia = valor_por_venda_dia['Valor Final'].mean()
    #print(ticket_medio_dia)
    
    if faturamento_dia >= meta_faturamento_dia:
        cor_fat_dia = 'green'
    else:
        cor_fat_dia = 'red'
    if faturamento_ano >= meta_faturamento_ano:
        cor_fat_ano = 'green'
    else:
        cor_fat_ano = 'red'
    if qte_produtos_dia >= meta_qteprodutos_dia:
        cor_qte_dia = 'green'
    else:
        cor_qte_dia = 'red'
    if qte_produtos_ano >= meta_qteprodutos_ano:
        cor_qte_ano = 'green'
    else:
        cor_qte_ano = 'red'
    if ticket_medio_dia >= meta_ticketmedio_dia:
        cor_ticket_dia = 'green'
    else:
        cor_ticket_dia = 'red'
    if ticket_medio_ano >= meta_ticketmedio_ano:
        cor_ticket_ano = 'green'
    else:
        cor_ticket_ano = 'red'

    nome = emails.loc[emails['Loja']==loja, 'Gerente'].values[0] # nome da pessoa. O values[0] traz o unico resultado em formato
    # de string, nao em tabela.
    
    Texto = f'''
    <p>Bom dia, {nome}!</p>

    <p>O resultado de ontem <strong>({dia_indicador.day}/{dia_indicador.month})</strong> da <strong>loja {loja}</strong> foi: </p>

        <table>
          <tr>
            <th>Indicador</th>
            <th>Valor Dia</th>
            <th>Meta Dia</th>
            <th>Cenário Dia</th>
          </tr>
          <tr>
            <td>Faturamento</td>
            <td style="text-align: center">R${faturamento_dia:.2f}</td>
            <td style="text-align: center">R${meta_faturamento_dia:.2f}</td>
            <td style="text-align: center"><font color="{cor_fat_dia}">◙</font></td>
          </tr>
          <tr>
            <td>Diversidade de Produtos</td>
            <td style="text-align: center">{qte_produtos_dia}</td>
            <td style="text-align: center">{meta_qteprodutos_dia}</td>
            <td style="text-align: center"><font color="{cor_qte_dia}">◙</font></td>
          </tr>
          <tr>
            <td>Ticket Médio</td>
            <td style="text-align: center">R${ticket_medio_dia:.2f}</td>
            <td style="text-align: center">R${meta_ticketmedio_dia:.2f}</td>
            <td style="text-align: center"><font color="{cor_ticket_dia}">◙</font></td>
          </tr>
        </table>
        <br>
        <br>
        <table>
          <tr>
            <th>Indicador</th>
            <th>Valor Ano</th>
            <th>Meta Ano</th>
            <th>Cenário Ano</th>
          </tr>
          <tr>
            <td>Faturamento</td>
            <td style="text-align: center">R${faturamento_ano:.2f}</td>
            <td style="text-align: center">R${meta_faturamento_ano:.2f}</td>
            <td style="text-align: center"><font color="{cor_fat_ano}">◙</font></td>
          </tr>
          <tr>
            <td>Diversidade de Produtos</td>
            <td style="text-align: center">{qte_produtos_ano}</td>
            <td style="text-align: center">{meta_qteprodutos_ano}</td>
            <td style="text-align: center"><font color="{cor_qte_ano}">◙</font></td>
          </tr>
          <tr>
            <td>Ticket Médio</td>
            <td style="text-align: center">R${ticket_medio_ano:.2f}</td>
            <td style="text-align: center">R${meta_ticketmedio_ano:.2f}</td>
            <td style="text-align: center"><font color="{cor_ticket_ano}">◙</font></td>
          </tr>
        </table>

    <p>Qualquer dúvida, estou à disposição.</p>
    <p>Att., Paulo</p>
    <br>
    <br>
    ''' 
    
    display(HTML(Texto))
```

```python
# ranking para diretoria:

# ranking anual:
faturamento_lojas = vendas.groupby('Loja')[['Loja', 'Valor Final']].sum() # filtra a tabela por loja e soma os valores. Mostra 2
# colunas, 'Loja' e 'Valor Final'
faturamento_lojas_ano = faturamento_lojas.sort_values(by='Valor Final', ascending=False) # ordena a tabela filtrada = descendente
print('Ranking faturamento anual:')
display(faturamento_lojas_ano)
# cria-se um arquivo excel
nome_arquivo = '{}_{}_Ranking Anual.xlsx'.format(dia_indicador.month, dia_indicador.day)
faturamento_lojas_ano.to_excel(r'Backup Arquivos Lojas\{}'.format(nome_arquivo))

# ranking diario
vendas_dia = vendas.loc[vendas['Data']==dia_indicador, :]
faturamento_lojas_dia = vendas_dia.groupby('Loja')[['Loja', 'Valor Final']].sum()
faturamento_lojas_dia = faturamento_lojas_dia.sort_values(by='Valor Final', ascending=False)
print('Ranking faturamento do dia:')
display(faturamento_lojas_dia)

nome_arquivo = '{}_{}_Ranking Dia.xlsx'.format(dia_indicador.month, dia_indicador.day)
faturamento_lojas_dia.to_excel(r'Backup Arquivos Lojas\{}'.format(nome_arquivo))
```
